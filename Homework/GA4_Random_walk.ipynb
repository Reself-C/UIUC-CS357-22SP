```python
import numpy as np
import matplotlib.pyplot as plt
def dice():
    return np.random.randint(1,7)
def flip():
    return np.random.choice([1,-1])
def step():
    return flip()*dice()
N = 1000
p0 = 100
```

```python
def simulate_asset_price(N,p0):
    return p0 + np.cumsum(np.array([0 if i == 0 else step() for i in range(N)]))

data = simulate_asset_price(N,p0)
import matplotlib.pyplot as plt
plt.plot(data)
```
