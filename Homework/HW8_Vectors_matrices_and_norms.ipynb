```python
# 8.11 Norm Ball
import matplotlib.pyplot as plt
import numpy as np

def norm(x, p):
    return np.sum(np.abs(x)**p, axis=0)**(1/p)

def plot_norm_ball(p):
    phi = np.linspace(0, 2 * np.pi, 500)
    unit_circle = np.array([np.cos(phi), np.sin(phi)])
    norm_ball = unit_circle/norm(unit_circle, p)
    plt.figure()
    plt.gca().set_aspect("equal")
    plt.plot(norm_ball[0], norm_ball[1])
    plt.xlim([-1.5, 1.5])
    plt.ylim([-1.5, 1.5])
    return plt.gca()

ps = [1, 2, 5, 0.5]
fig_1, fig_2, fig_3, fig_4 = [plot_norm_ball(p) for p in ps]
```
